```python
import sqlite3
import pandas as pd
```

```python
# Load CSV file into a Pandas DataFrame
df = pd.read_csv('infant_mortality.csv', encoding='latin-1')
```

```python
# Create a connection to a new SQLite database
conn = sqlite3.connect('infant_mortality.db')
```

```python
# Write the DataFrame to a new SQLite table called "infant_mortality"
df.to_sql('infant_mortality', conn, if_exists='replace', index=False)
```

```python
# Close the connection
conn.close()
```

```python
from sqlalchemy import create_engine

engine = create_engine("sqlite:///infant_mortality.db")
```

```python
conn = engine.connect()
df = pd.read_sql('Select * From infant_mortality', conn)
data = df.to_dict(orient="records")
```
